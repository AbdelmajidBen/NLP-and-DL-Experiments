```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

# URL of the website to scrape
base_url = "https://arabic.cnn.com/"
url = "https://arabic.cnn.com/"

# Send a GET request to the webpage
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Find all <a> tags containing links
    links = soup.find_all('a', href=True)
    
    # Initialize a counter for processed links
    count = 0
    
    # Process the links
    for link in links:
        # Check if the link is not a social media link
        if "twitter.com" not in link['href'] and "facebook.com" not in link['href'] and "instagram.com" not in link['href']:
            # Prepend the base URL to the relative link
            absolute_link = urljoin(base_url, link['href'])
            
            # Send a GET request to the link
            link_response = requests.get(absolute_link)
            
            # Check if the request to the link was successful
            if link_response.status_code == 200:
                # Parse the HTML content of the link page
                link_soup = BeautifulSoup(link_response.content, 'html.parser')
                
                # Find all <p> tags within the link page
                paragraphs = link_soup.find_all('p')
                
                # Print the content of <p> tags
                print(f"Content from link: {absolute_link}")
                for paragraph in paragraphs:
                    print(paragraph.text.strip())
                print("=" * 50)  # Separate the content from different links
                
                # Increment the counter
                count += 1
                
                # Check if we have processed 10 links
                if count == 20:
                    break
            else:
                print(f"Failed to retrieve content from link: {absolute_link}. Status code: {link_response.status_code}")
else:
    print(f"Failed to retrieve the webpage. Status code: {response.status_code}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL to scrape
url = "https://www.oracle.com/ae-ar/artificial-intelligence/what-is-ai/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, "html.parser")

# Find all <h2> and <p> elements on the page
headings = soup.find_all(["h2", "p"])

# Extract text from <h2> and <p> elements and create a list of sentences with a maximum of 10 words
sentences = []
for tag in headings:
    text = tag.text.strip()
    # Split text into sentences
    text_sentences = text.split('.')
    # Iterate through sentences and keep only those with a maximum of 10 words
    for sentence in text_sentences:
        # Remove leading and trailing whitespace
        sentence = sentence.strip()
        # Split sentence into words
        words = sentence.split()
        if len(words) <= 10 and sentence:
            sentences.append(sentence)

# Create a DataFrame with two columns: text and score (initially empty)
data = {"text": sentences, "score": [""] * len(sentences)}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("text_data_with_score1.csv", index=False)

print("Data saved to text_data_with_score.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# URL to scrape
url = "https://www.oracle.com/ae-ar/artificial-intelligence/what-is-ai/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, "html.parser")

# Find the main <article> tag
article = soup.find("article")

# Extract text from the main content
text = article.get_text() if article else ""

# Split the text into sentences with a maximum of 10 words each
sentences = re.findall(r"[\w\s'-]{1,100}", text)

# Create a DataFrame with two columns: text and score (initially empty)
data = {"text": sentences, "score": [""] * len(sentences)}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("text_data_with_score.csv", index=False)

print("Data saved to text_data_with_score.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# URL to scrape
url = "https://www.oracle.com/ae-ar/artificial-intelligence/what-is-ai/"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, "html.parser")

# Find all <h2> and <p> elements on the page
headings = soup.find_all(["h2", "p"])

# Extract full text from <h2> and <p> elements
full_text = ""
for tag in headings:
    full_text += tag.text.strip() + " "

# Split full text into words
words = re.findall(r'\b\w+\b', full_text)

# Initialize an empty list to hold sentences with a maximum of 5 words
sentences = []
sentence = ""
# Iterate over words and form sentences with a maximum of 5 words
for word in words:
    if len(sentence.split()) < 5:
        sentence += word + " "
    else:
        sentences.append(sentence.strip())
        sentence = word + " "
# Append the last sentence
if sentence:
    sentences.append(sentence.strip())

# Create a DataFrame with two columns: text and score (initially empty)
data = {"text": sentences, "score": [""] * len(sentences)}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("text_data_with_score.csv", index=False)

print("Data saved to text_data_with_score.csv")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

# URL to scrape
url = "https://www.aljazeera.net/encyclopedia/2016/6/6/%D8%A7%D9%84%D8%B0%D9%83%D8%A7%D8%A1-%D8%A7%D9%84%D8%A7%D8%B5%D8%B7%D9%86%D8%A7%D8%B9%D9%8A-%D8%AD%D9%8A%D9%86%D9%85%D8%A7-%D8%AA%D9%81%D9%83%D8%B1-%D8%A7%D9%84%D8%A2%D9%84%D8%A9"

# Send a GET request to the URL
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.content, "html.parser")

# Find all <p> elements on the page
paragraphs = soup.find_all("p")

# Extract full text from <p> elements
full_text = ""
for p in paragraphs:
    full_text += p.get_text(strip=True) + " "

# Split full text into words
words = re.findall(r'\b\w+\b', full_text)

# Initialize an empty list to hold sentences with a maximum of 5 words
sentences = []
sentence = ""
# Iterate over words and form sentences with a maximum of 5 words
for word in words:
    if len(sentence.split()) < 5:
        sentence += word + " "
    else:
        sentences.append(sentence.strip())
        sentence = word + " "
# Append the last sentence
if sentence:
    sentences.append(sentence.strip())

# Create a DataFrame with two columns: text and score (initially empty)
data = {"text": sentences, "score": [""] * len(sentences)}
df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
df.to_csv("text_data_from_aljazeera_website.csv", index=False)

print("Data saved to text_data_from_aljazeera_website.csv")
```
